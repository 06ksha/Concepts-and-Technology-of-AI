```python
#Problem_1:

import numpy as np

# 1. Empty array 2x2
arr_empty = np.empty((2, 2))
print("Empty 2x2 array:\n", arr_empty)

# 2. All-one array 4x2
arr_ones = np.ones((4, 2))
print("\nAll ones 4x2 array:\n", arr_ones)

# 3. Array filled with a value (example: 7)
arr_full = np.full((3, 3), 7)
print("\nFull array filled with 7:\n", arr_full)

# 4. Zeros with same shape/type as another array
a = np.array([[1, 2], [3, 4]])
arr_zeros_like = np.zeros_like(a)
print("\nZeros like array a:\n", arr_zeros_like)

# 5. Ones with same shape/type as another array
arr_ones_like = np.ones_like(a)
print("\nOnes like array a:\n", arr_ones_like)

# 6. Convert list to numpy array
new_list = [1, 2, 3, 4]
arr_from_list = np.array(new_list)
print("\nArray from list:\n", arr_from_list)
```

```python
#Problem_2:

# 1. Array from 10 to 49
arr_range = np.arange(10, 50)
print("\nArray 10 to 49:\n", arr_range)

# 2. 3x3 matrix from 0 to 8
arr_matrix = np.arange(9).reshape(3, 3)
print("\n3x3 matrix 0 to 8:\n", arr_matrix)

# 3. Identity matrix 3x3
identity_matrix = np.eye(3)
print("\nIdentity matrix:\n", identity_matrix)

# 4. Random array of size 30 & mean
rand_arr = np.random.random(30)
print("\nMean of random array:", rand_arr.mean())

# 5. 10x10 random array: min & max
rand_10 = np.random.random((10, 10))
print("\nMin:", rand_10.min(), "Max:", rand_10.max())

# 6. Zero array with 5th element = 1
zero_arr = np.zeros(10)
zero_arr[4] = 1
print("\nZero array with 5th element 1:\n", zero_arr)

# 7. Reverse array
arr = np.array([1, 2, 0, 0, 4, 0])
rev_arr = arr[::-1]
print("\nReversed array:\n", rev_arr)

# 8. 2D array with border = 1, inside = 0
border_arr = np.ones((5, 5))
border_arr[1:-1, 1:-1] = 0
print("\n2D border matrix:\n", border_arr)

# 9. 8x8 checkerboard pattern
checkerboard = np.zeros((8, 8))
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n8x8 checkerboard:\n", checkerboard)
```

```python
#Problem_4:

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
identity_check = A.dot(A_inv)
print("\nA * A^-1:\n", identity_check)

# 2. Prove AB ≠ BA
AB = A.dot(B)
BA = B.dot(A)
print("\nAB:\n", AB)
print("\nBA:\n", BA)

# 3. Prove (AB)^T = B^T A^T
lhs = AB.T
rhs = B.T.dot(A.T)
print("\n(AB)^T:\n", lhs)
print("\nB^T A^T:\n", rhs)
```

```python
#Write as matrices AX = B
A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

B = np.array([-1, -3, 9])

#Solve using inverse method
A_inv = np.linalg.inv(A)
X = A_inv.dot(B)
print("\nSolution using inverse method:\n", X)

#Solve using NumPy's direct solver
X2 = np.linalg.solve(A, B)
print("\nSolution using np.linalg.solve:\n", X2)
```
